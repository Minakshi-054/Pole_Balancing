```python

import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'Precision':.75, 'Recall':.88, 'F1 Score':.81
        }
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses[0], values[0], color ='maroon',
        width = 0.4)
plt.bar(courses[1], values[1], color ='blue',
        width = 0.4)
plt.bar(courses[2], values[1], color ='grey',
        width = 0.4)

 
plt.xlabel("Result on Simulated validation set")
plt.ylabel("Value")
#plt.title("Students enrolled in different courses")
plt.show()
```
